Question 1: What is Simple Linear Regression?

Answer:**Simple Linear Regression** is a statistical technique used to examine and model the relationship between two variables: one independent variable (predictor) and one dependent variable (response). It aims to find a straight-line equation that best predicts the dependent variable based on the independent variable. The relationship is expressed in the form (y = b_0 + b_1 x + \varepsilon), where (y) represents the dependent variable, (x) is the independent variable, (b_0) is the intercept, (b_1) is the slope indicating how much (y) changes for a one-unit change in (x), and (\varepsilon) is the error term accounting for the difference between observed and predicted values. Simple Linear Regression assumes a linear relationship between the two variables, and the line of best fit is determined by minimizing the sum of squared errors using the method of ordinary least squares. This technique is commonly used both for prediction purposes and for understanding the strength and direction of the relationship between variables.


Question 2: What are the key assumptions of Simple Linear Regression?

Answer:The key assumptions of **Simple Linear Regression** are crucial for ensuring that the model provides valid and reliable results. First, it assumes a **linear relationship** between the independent variable and the dependent variable, meaning changes in the predictor correspond to proportional changes in the response. Second, the **errors (residuals) are normally distributed**, which allows for accurate hypothesis testing and confidence intervals. Third, the errors should have **constant variance** across all levels of the independent variable, a property known as **homoscedasticity**, meaning the spread of residuals does not change with the predictor. Fourth, the errors are assumed to be **independent**, implying that the residual for one observation does not influence another. Finally, the independent variable is assumed to be measured **without error**, and there should be no perfect **multicollinearity**, though in simple linear regression with one predictor, this is inherently satisfied. These assumptions collectively ensure that the regression estimates are unbiased, efficient, and statistically valid.


Question 3: What is heteroscedasticity, and why is it important to address in regression
models?

Answer:Heteroscedasticity refers to a situation in regression analysis where the variance of the errors (residuals) is not constant across all levels of the independent variable. In other words, as the value of the predictor changes, the spread of the residuals either increases or decreases instead of remaining uniform. This violates one of the key assumptions of linear regression, which assumes homoscedasticity, or constant variance of errors.

Addressing heteroscedasticity is important because it can invalidate the results of the regression model. When heteroscedasticity is present, the estimates of the regression coefficients remain unbiased, but the standard errors become unreliable, which affects hypothesis tests, confidence intervals, and the overall inference about the relationships between variables. This can lead to incorrect conclusions, such as thinking a variable is significant when it is not, or vice versa. Detecting heteroscedasticity and applying remedies—such as transforming variables, using weighted least squares, or robust standard errors—helps ensure that the regression model produces valid and trustworthy results.

Question 4: What is Multiple Linear Regression?

Answer:**Multiple Linear Regression** is an extension of simple linear regression that models the relationship between a **dependent variable** and **two or more independent variables**. It aims to predict the value of the dependent variable based on the combined effect of several predictors, rather than just one. The general form of the multiple linear regression equation is (y = b_0 + b_1 x_1 + b_2 x_2 + \dots + b_n x_n + \varepsilon), where (y) is the dependent variable, (x_1, x_2, \dots, x_n) are the independent variables, (b_0) is the intercept, (b_1, b_2, \dots, b_n) are the coefficients representing the change in (y) for a one-unit change in each predictor while keeping others constant, and (\varepsilon) is the error term. Multiple linear regression assumes a **linear relationship** between the dependent variable and all independent variables, and it is widely used for **prediction, analysis of relationships, and understanding the relative importance of different factors** influencing the outcome.


Question 5: What is polynomial regression, and how does it differ from linear
regression?

Answer:**Polynomial Regression** is a type of regression analysis used to model the relationship between a dependent variable and an independent variable when the relationship is **non-linear**. Instead of fitting a straight line as in linear regression, polynomial regression fits a **curved line** by including higher-degree terms of the independent variable, such as (x^2, x^3), etc. The general form of a polynomial regression equation is (y = b_0 + b_1 x + b_2 x^2 + \dots + b_n x^n + \varepsilon), where (y) is the dependent variable, (x) is the independent variable, (b_0, b_1, \dots, b_n) are coefficients, and (\varepsilon) is the error term.

The key difference between polynomial and simple linear regression is that linear regression assumes a **straight-line relationship** between the independent and dependent variables, whereas polynomial regression can model **curved, more complex relationships**. This makes polynomial regression more flexible for capturing trends in data that are not adequately represented by a straight line, but it also requires careful selection of the polynomial degree to avoid **overfitting**.


Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:

● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points.

(Include your Python code and output in the code box below.)

Answer:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

slope = model.coef_[0]
intercept = model.intercept_
print(f"Regression equation: Y = {intercept:.2f} + {slope:.2f}X")

Y_pred = model.predict(X)

plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

Question 7: Fit a Multiple Linear Regression model on this sample data:

● Area = [1200, 1500, 1800, 2000]

● Rooms = [2, 3, 3, 4]

● Price = [250000, 300000, 320000, 370000]

Check for multicollinearity using VIF and report the results.

(Include your Python code and output in the code box below.)

Answer:


```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

data = pd.DataFrame({
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
})

X = data[['Area', 'Rooms']]
y = data['Price']

model = LinearRegression()
model.fit(X, y)

print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {dict(zip(X.columns, model.coef_))}")

X_with_const = sm.add_constant(X)
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_with_const.values, i+1) for i in range(len(X.columns))]

print("\nVariance Inflation Factor (VIF) for each feature:")
print(vif_data)
```

Question 8: Implement polynomial regression on the following data:

● X = [1, 2, 3, 4, 5]
3

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.

(Include your Python code and output in the code box below.)

Answer:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

Y_pred = model.predict(X_poly)

plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Polynomial regression (degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.show()

print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

Question 9: Create a residuals plot for a regression model trained on this data:

● X = [10, 20, 30, 40, 50]

● Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.

(Include your Python code and output in the code box below.)

Answer:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)
residuals = Y - Y_pred

plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()

print("Residuals:", residuals)
```

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

Answer: As a data scientist encountering **heteroscedasticity** and **multicollinearity** while predicting house prices, I would take several steps to address these issues and ensure a robust regression model.

To address **heteroscedasticity**, I would first visualize the residuals using a residuals plot to confirm the pattern of non-constant variance. If heteroscedasticity is present, I could apply **transformations** to the dependent variable, such as a logarithmic, square root, or Box-Cox transformation, which often stabilizes variance. Alternatively, I could use **weighted least squares regression** or **robust standard errors**, which adjust the model to account for unequal variance in the errors, ensuring that coefficient estimates remain valid and reliable.

For **multicollinearity**, which occurs when independent variables are highly correlated, I would start by calculating **Variance Inflation Factors (VIFs)** to identify problematic predictors. To reduce multicollinearity, I could **remove or combine highly correlated variables**, apply **dimensionality reduction techniques** such as Principal Component Analysis (PCA), or use **regularization methods** like Ridge or Lasso regression that penalize large coefficients and reduce the impact of collinear predictors.

Finally, I would **validate the model** using cross-validation or a separate test set to ensure its predictive performance is robust and generalizes well. By combining these steps—transforming variables, adjusting for heteroscedasticity, managing multicollinearity, and thorough validation—I can develop a reliable regression model that produces accurate and interpretable house price predictions.
