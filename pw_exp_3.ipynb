1. What is the difference between a function and a method in Python?

ANS-In Python, a function is a block of reusable code that performs a specific task and is defined independently using the def keyword. It can be called by its name and does not belong to any object. For example, a user-defined function like def greet(name): return f"Hello, {name}!" can be called with greet("Alice") to return "Hello, Alice!". On the other hand, a method is a function that is associated with an object and is called using the dot (.) notation on that object. Methods are defined within a class and operate on instances of that class. For instance, the string method upper() can be used as "alice".upper() to return "ALICE". In summary, while both functions and methods perform specific tasks, the key difference lies in their association—functions are standalone, whereas methods are tied to objects.

2. Explain the concept of function arguments and parameters in Python.

ANS-In Python, parameters and arguments are terms used in the context of functions, and while they're closely related, they refer to different things. Parameters are the variable names listed in a function’s definition—they act as placeholders for the values the function expects to receive. Arguments, on the other hand, are the actual values passed to a function when it is called.

For example, in the function definition def add(a, b):, a and b are parameters. When the function is called like add(3, 5), the values 3 and 5 are the arguments. Python supports several types of arguments: positional arguments, where the order matters; keyword arguments, where you explicitly specify which parameter gets which value (e.g., add(a=3, b=5)); default arguments, where parameters have default values if no argument is provided; and arbitrary arguments using *args and **kwargs to accept variable numbers of arguments. This flexibility makes Python functions powerful and versatile.

3. What are the different ways to define and call a function in Python?

ANS-In Python, functions can be defined and called in several ways to suit different programming needs. The most common way to define a function is by using the def keyword. For example, def greet(name): return f"Hello, {name}!" defines a function named greet that takes one parameter. A function can also be defined using a lambda expression, which creates small anonymous functions. For instance, add = lambda x, y: x + y defines a function that adds two numbers. Functions can also be nested inside other functions for encapsulation. For example, inside def outer():, you might have def inner(): print("Hi"), and calling outer() will run the nested inner().

When it comes to calling functions, there are several options. Positional arguments are the most straightforward, where the values are passed in the same order as parameters—e.g., greet("Alice"). With keyword arguments, you can specify the parameter names—e.g., greet(name="Bob"). You can also define default arguments in the function—like def greet(name="Guest"):—which allows calling greet() without passing any value, returning "Hello, Guest!". For functions that need to accept multiple arguments, Python allows variable-length arguments using *args and **kwargs. For example, def show(*args): print(args) can take any number of positional values like show(1, 2, 3), and def info(**kwargs): print(kwargs) accepts any number of keyword arguments like info(name="Alice", age=25). These different ways to define and call functions give Python great flexibility in structuring code.

5. What are iterators in Python and how do they differ from iterables?

ANS-In Python, an iterator is an object that represents a stream of data and allows you to access elements one at a time using the next() function. An iterable, on the other hand, is any Python object capable of returning its elements one at a time, typically using a loop, but it does not necessarily maintain its own iteration state.

An iterable is any object that implements the __iter__() method, such as lists, tuples, strings, and dictionaries. You can loop over an iterable using a for loop. However, to get an iterator from it, you must explicitly call iter() on it. The resulting iterator object implements both __iter__() and __next__() methods. The __next__() method fetches the next item from the sequence and raises a StopIteration exception when there are no more items.

numbers = [1, 2, 3]

it = iter(numbers)

print(next(it))

print(next(it))

print(next(it))

In this example, numbers is an iterable, and it is the iterator that retrieves each element using next(). Once all items are consumed, calling next() again would raise a StopIteration exception.


6. Explain the concept of generators in Python and how they are defined.

ANS-In Python, generators are a special type of iterable that allow you to iterate over a sequence of values lazily, meaning they produce values one at a time and only when needed, rather than storing the entire sequence in memory. This makes them very efficient for working with large data sets or streams of data.

Generators are defined using functions with the yield keyword instead of return. When a function contains yield, it becomes a generator function and returns a generator object. Each time the generator's __next__() method is called (or used in a for loop), the function resumes from where it left off and produces the next value.


def count_up_to(n):

    i = 1

    while i <= n:

        yield i

        i += 1

counter = count_up_to(3)

print(next(counter))  

print(next(counter))  

print(next(counter))  

In this example, count_up_to is a generator function that yields numbers from 1 to n. Each call to next() retrieves the next value, and the state is preserved between calls. Generators are memory-efficient and ideal for looping through large or infinite sequences.

7. What are the advantages of using generators over regular functions?

ANS-Generators offer several advantages over regular functions, especially when working with large data sets or streams. The key benefits include:

Memory Efficiency: Generators produce values one at a time using the yield keyword, so they don’t store the entire result in memory. This is especially useful when dealing with large sequences or files.

Lazy Evaluation: Generators compute values only when requested (on-the-fly), which makes programs faster and more responsive, especially when not all values are needed at once.

Improved Performance: Because generators avoid the overhead of storing data in lists or other structures, they often run faster in scenarios where data can be processed one item at a time.

Cleaner Code: Generator functions are typically shorter and more readable than functions that return large data collections. They use simple loops and yield instead of managing lists manually.

State Preservation: Generators automatically maintain their state between calls, making them ideal for implementing iterators without writing boilerplate code.

def generate_numbers():
    for i in range(1, 1000000):
        yield i

In this example, generate_numbers() can handle a huge range of numbers efficiently without ever loading them all into memory. This wouldn't be practical with a regular function returning a list of a million items.

8. What is a lambda function in Python and when is it typically used?

ANS-A lambda function in Python is an anonymous, one-line function defined using the lambda keyword instead of the usual def. It can take any number of arguments but can only contain a single expression, which is evaluated and returned. Lambda functions are typically used for short, simple operations where writing a full function would be unnecessarily verbose. They are especially useful in situations where functions are used temporarily, such as within built-in functions like map(), filter(), or sorted(). For example, lambda x, y: x + y defines a function that adds two numbers and can be used as add = lambda x, y: x + y followed by add(3, 5) to return 8. In another case, using map(lambda x: x**2, [1, 2, 3]) will return a list of squared numbers. Lambda functions help keep code concise and are ideal for simple, one-time functional logic.

9. Explain the purpose and usage of the `map()` function in Python.

ANS-The map() function in Python is used to apply a specific function to each item in an iterable, such as a list or tuple, and returns an iterator containing the results. It is particularly useful when you want to transform or process all elements of a sequence in the same way without writing an explicit loop. The map() function takes two arguments: the function to apply and the iterable to process. For example, map(lambda x: x**2, [1, 2, 3, 4]) applies a lambda function that squares each number in the list, resulting in [1, 4, 9, 16] when converted to a list. This approach helps make code cleaner and more concise, especially when performing simple operations on collections of data.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

ANS-In Python, map(), reduce(), and filter() are built-in functions used for functional programming, but each serves a different purpose.

The map() function is used to apply a function to every item in an iterable and returns a new iterator with the results. It transforms each element individually. For example, map(lambda x: x * 2, [1, 2, 3]) returns [2, 4, 6].

The filter() function is used to select items from an iterable based on a condition. It applies a function that returns True or False to each item and only keeps those for which the function returns True. For example, filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) returns [2, 4], filtering out the odd numbers.

The reduce() function, which is available in the functools module, is used to repeatedly apply a function to pairs of items in an iterable, reducing the iterable to a single value. For example, reduce(lambda x, y: x + y, [1, 2, 3, 4]) returns 10, as it sums the list step by step.

In summary, map() transforms, filter() selects, and reduce() combines, each serving a unique and powerful role in processing data functionally.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

ANS-

```python
from functools import reduce

def sum_of_even_numbers(numbers):
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return reduce(lambda x, y: x + y, even_numbers, 0)

example_list = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(example_list)
print("Sum of even numbers:", result)
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

ANS-

```python
from functools import reduce

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.sequence = list(map(lambda i: reduce(lambda a, _: (a[1], a[0] + a[1]), range(i), (0, 1))[0], range(max_terms)))
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.max_terms:
            raise StopIteration
        val = self.sequence[self.index]
        self.index += 1
        return val

n = int(input("Enter number of Fibonacci terms: "))
fib_iter = FibonacciIterator(n)

for num in fib_iter:
    print(num, end=' ')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

ANS-

```python
def powers_of_two(max_exp):
    for x in range(max_exp + 1):
        yield 2 ** x

n = int(input("Enter the maximum exponent: "))

for val in powers_of_two(n):
    print(val, end=' ')
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

ANS-

```python
def read_lines(filepath):
    with open(filepath, 'r') as file:
        yield from map(lambda line: line.rstrip('\n'), file)

filename = input("Enter the file path: ")

for line in read_lines(filename):
    print(line)
```
